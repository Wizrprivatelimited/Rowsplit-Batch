```python
df1

# Replace 'Data' with the actual column name you want to process
column_to_split = "PID"  # Replace with the name of your column

# Initialize a new DataFrame to store the results
processed_data = pd.DataFrame()

try:
    # Process the data from bottom to top
    for index in reversed(df1.index):
        # Get the current cell value
        cell_value = df1.at[index, column_to_split]

        if pd.isna(cell_value) or cell_value == "":
            # If the cell is empty, add it as is
            processed_data = pd.concat([processed_data, df1.iloc[[index]]])
        else:
            # Split the cell value into a list if it contains data
            split_values = str(cell_value).split(",")  # Update delimiter if needed
            for value in split_values:
                # Append each split value as a new row, duplicating other columns
                row = df1.iloc[[index]].copy()
                row[column_to_split] = value.strip()  # Remove extra spaces
                processed_data = pd.concat([processed_data, row])
except Exception as e:
    print(f"An error occurred: {e}")

# Reverse the processed DataFrame to maintain the original order
processed_data = processed_data.iloc[::-1]

# Save the result to a new Excel file
output_file = "/Users/romitbenkar/Downloads/Split_data.xlsx"
processed_data.to_excel(output_file, index=False)

print(f"Data has been processed and saved to {output_file}")
```
