```python
## both concept and function 

import pandas as pd

# Load your file
file_path = r"C:\Users\taslim.siddiqui\Downloads\ABC Prod data Product-Skill-Concept1.xlsx"
df = pd.read_excel(file_path)

# Drop rows where either 'concepts' or 'function_ids' is missing
df = df.dropna(subset=['concepts', 'function_ids'])

# Convert to string and split by comma
df['concepts'] = df['concepts'].astype(str).str.split(r'\s*,\s*')
df['function_ids'] = df['function_ids'].astype(str).str.split(r'\s*,\s*')

# Explode both columns (creates all combinations)
df = df.explode('concepts')
df = df.explode('function_ids').reset_index(drop=True)

# Remove blank or whitespace-only values
df = df[df['concepts'].str.strip().astype(bool)]
df = df[df['function_ids'].str.strip().astype(bool)]

# Reorder to show concepts and function_ids first
cols = ['concepts', 'function_ids'] + [col for col in df.columns if col not in ['concepts', 'function_ids']]
df = df[cols]

# Save to Excel
output_path = "C:\\Users\\taslim.siddiqui\\Downloads\\Split_Concepts_Functions_ABC_prod.xlsx"
df.to_excel(output_path, index=False)

print(f"✅ File saved to: {output_path}")


```

```python
import pandas as pd

# Load your file
file_path = r"C:\Users\taslim.siddiqui\Downloads\Master_FSC_Sheet_ABC&Sarvodaya New 20.03.xlsx"
df = pd.read_excel(file_path)

# Drop rows where Company is missing
df = df.dropna(subset=['Company'])

# Convert to string and split by comma
df['Company'] = df['Company'].astype(str).str.split(',')

# Explode (split into multiple rows)
df = df.explode('Company').reset_index(drop=True)

# Clean spaces (VERY IMPORTANT)
df['Company'] = df['Company'].str.strip()

# Remove empty values
df = df[df['Company'] != '']

# Save to Excel
output_path = r"C:\Users\taslim.siddiqui\Downloads\Company_split.xlsx"
df.to_excel(output_path, index=False)

print(f"✅ File saved to: {output_path}")
```
